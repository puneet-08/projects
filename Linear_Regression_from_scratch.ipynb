**Linear Regression**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Generating Random Data for Regression

```python
from sklearn.datasets import make_regression
X, y = make_regression(n_samples = 200, n_features = 1, n_informative=1, noise = 6, bias =30, random_state= 200)
y=y.reshape(-1,1)
```

Visualizing the Data

```python
plt.scatter(X,y, c ='red', alpha =.5, marker = 'o')
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```

Linear Model

```python
def h(X,w):
  return w[1]*np.array(X) + w[0]
```

Cost

```python
def cost(w,X,y):
  return (np.sum(np.square(h(X,w) - y)))/(2*X.shape[0])
```

Partial derivatives

```python
def grad(w,X,y):
  g = [0]*2
  g[0] = np.sum(h(X,w)-y)/(X.shape[0])
  g[1] = np.sum((h(X,w)-y)*X)/(X.shape[0])

  return np.array(g)
```

Grad Descent

```python
def grad_dec(w,X,y,lr=0.01, n_iter =300):
  plt.scatter(X,y, c ='red', alpha =.5, marker = 'o')
  cost_ = []
  w_ =  []
  for i in range(n_iter):
    linear_model = h(X,w)
    cost_.append(cost(w,X,y))
    w = w -lr*grad(w,X,y)
    w_.append(w)

    if (i%30 ==0) or (i==1):
      plt.plot(np.linspace(-3, 3), h(np.linspace(-3, 3),w), c = 'blue',alpha = 0.5)

  plt.xlabel('X')
  plt.ylabel('y')

  return plt.show(), cost_

```

```python
#Initializing weight matrix
w = [0,0]
```

```python
#Plotting Gradient Descent and journey of the fitted line
_ , cost = grad_dec(w,X,y)
```

```python
#Plotting Cost function w.r.t. iterations
plt.plot(cost)
plt.xlabel('iterations')
plt.ylabel('cost')
plt.show()
```
